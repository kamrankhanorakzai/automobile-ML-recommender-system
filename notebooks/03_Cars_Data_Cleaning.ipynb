```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"..\data\raw\cars_data.csv")
df = df.rename(columns={'kolimeter': 'kilometer', 'feul_type': 'fuel_type'})
df
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.info()
```

```python
df['city'] = df['city'].str.lower().str.strip()
df.value_counts('city').sort_index()
```

```python
df['price'] = (
    df['price']
    .str.replace(r'[^0-9.]', '', regex=True)
)

df['price'] = pd.to_numeric(df['price'])* 100000
```

```python
df
```

```python
df['price'].isna().sum()
```

```python
df['kilometer'] = (
    df['kilometer']
    .str.replace(r'[^0-9]', '', regex=True)
)
df['kilometer'] = pd.to_numeric(df['kilometer'])
```

```python
df['engine'] = (
    df['engine']
    .str.replace(r'[^0-9]', '', regex=True)
)
df['engine'] = pd.to_numeric(df['engine'])
```

```python
df["fuel_type"]=df["fuel_type"].str.lower()
df["fuel_type"].value_counts()
```

```python
df["fuel_type"].str.replace("lp","cng")
```

```python
df["transmission"]=df["transmission"].str.lower()
df["transmission"].value_counts()
```

```python
df['brand'] = df['name'].str.lower().str.split().str[0]
```

```python
df['brand'].value_counts() 
```

```python
df['name'] = (
    df['name']
    .str.replace(r'\d{4}', '', regex=True)        # remove year
    .str.replace('for Sale', '', regex=False)     # remove noise text
    .str.replace(r'\s+', ' ', regex=True)         # remove extra spaces
    .str.strip()
    .str.split()
    .str.join(' ')
)
```

```python
df
```

```python
df.to_csv('../data/interim/cars_cleaned.csv', index=False)
```
