```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
with open("pakwheels.html", "r", encoding="utf-8") as f:
    soup = BeautifulSoup(f, "html.parser")
```

```python
cars = soup.find_all("h3")

name = []

for car in cars:
    name.append(car.text.strip())
```

```python
city_blocks = soup.find_all("ul", class_="list-unstyled search-vehicle-info fs13")
city = []

for block in city_blocks:
    c = block.find("li").text.strip()
    city.append(c)
```

```python
prices_data = soup.find_all("div", class_="price-details generic-dark-grey")
prices = []

for p in prices_data:
    price = p.text.strip().replace("\n", "").replace("  ", " ")
    prices.append(price)
```

```python
info_blocks = soup.find_all("ul", class_="list-unstyled search-vehicle-info-2 fs13")

model = []

for block in info_blocks:
    year = block.find_all("li")[0].text.strip()
    model.append(year)
```

```python
kms = []

for block in info_blocks:
    km = block.find_all("li")[1].text.strip()
    kms.append(km)
```

```python
fuels = []

for block in info_blocks:
    fuel = block.find_all("li")[2].text.strip()
    fuels.append(fuel)
```

```python
engine = []

for block in info_blocks:
    eng = block.find_all("li")[3].text.strip()
    engine.append(eng)
```

```python
transmissions = []

for block in info_blocks:
    transmission = block.find_all("li")[4].text.strip()
    transmissions.append(transmission)
    
```

```python
unwanted_strings = {
    'View Cars by City', 'View Cars by Make', 'View Cars by Model', 'View Cars by Version',
    'View Cars by Engine Capacity', 'View Cars by Category', 'View Cars by Year',
    'Enter Your Email', 'Let’s get you started!','Forgot your password?'
}
name = [car for car in name if car not in unwanted_strings]
len(name)
```

```python
# List of cars to remove
remove_cars = [
    "Suzuki Mehran 2001",
    "Hyundai Santa Fe 2024",
    "BMW 7 Series 2005",
    "MG HS 2025",
    "Honda Civic 2012",
    "Mazda Flair Crossover 2020",
    "Toyota Vitz 2003",
    "Toyota Aqua 2015",
    "Toyota Camry 2014",
    "Mercedes Benz S Class 2008",
    "Honda City 2020",
    "Toyota Hilux 2021"
]

# Remove them from 'name' and update 'name' itself
name = [car for car in name if car not in remove_cars]
len(name)
```

```python
df = pd.DataFrame({
    "name": name,
    "city": city,
    "price": prices,
    "model": model,
    "kolimeter": kms,
    "feul_type": fuels,
    "engine": engine,
    "transmission": transmissions
})
df
```

```python
df.to_csv("cars_data.csv", index=False)
```
